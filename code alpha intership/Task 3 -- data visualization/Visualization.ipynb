```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

# Load dataset
df = pd.read_csv("books.csv")

df.head()
```

```python
# Remove currency symbol and convert to float
df["Price"] = (
    df["Price"]
    .str.replace("Â£", "", regex=False)
    .str.replace("£", "", regex=False)
    .astype(float)
)

df.info()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x="Rating", data=df, order=["One","Two","Three","Four","Five"])
plt.title("Distribution of Book Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Books")
plt.show()
```

### Insight

- This chart shows the distribution of book ratings.
- Most books belong to the highest-frequency rating category, while fewer books have lower ratings.

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Price"], bins=8, kde=True)
plt.title("Distribution of Book Prices")
plt.xlabel("Price (£)")
plt.ylabel("Count")
plt.show()
```

### Insight

- Most book prices fall within a similar price range.
- Only a few books are significantly more expensive than the others.

```python
plt.figure(figsize=(6,5))
sns.boxplot(y=df["Price"])
plt.title("Book Price Box Plot")
plt.show()
```

### Insight

- The box plot shows the spread of book prices.
- It helps identify any unusually high or low-priced books (outliers).

```python
avg_price = df.groupby("Rating")["Price"].mean().reindex(
    ["One","Two","Three","Four","Five"]
)

plt.figure(figsize=(8,5))
avg_price.plot(kind="bar")
plt.title("Average Book Price by Rating")
plt.xlabel("Rating")
plt.ylabel("Average Price (£)")
plt.show()
```

### Insight

- This chart compares the average price of books across different rating categories.
- It helps determine whether highly rated books tend to be more expensive.
